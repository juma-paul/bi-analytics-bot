```python
essential_columns = [
    'Country',
    'Transportation Mode',
    'Product Category',
    'Agreed Delivery Date Year Month',
    'On Time (OTD)',
    'Average Days Late',
    'Number of Shipments'
]
```

```python
logistics_data = data[essential_columns].copy()
```

```python
logistics_data
```

```python
logistics_data.dtypes
```

```python
logistics_data["Agreed Delivery Date Year Month"] = pd.to_datetime(logistics_data["Agreed Delivery Date Year Month"], format="%Y-%m")
```

```python
logistics_data["On Time (OTD)"] = logistics_data["On Time (OTD)"].replace({'Y': 1, 'N': 0})
logistics_data["On Time (OTD)"] = logistics_data["On Time (OTD)"].astype(float)
logistics_data.dtypes
```

```python
logistics_data = logistics_data.dropna(subset=["Agreed Delivery Date Year Month"])
```

```python
logistics_data = logistics_data[logistics_data["Agreed Delivery Date Year Month"].dt.year <= 2025]
```

```python
logistics_data
```

```python
logistics_data.isnull().sum()
```

```python
logistics_data.dropna(inplace=True)
```
